```python
#Task_1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

#  
file_path = '_16_.xlsx'  #    
data = pd.read_excel(file_path)

#  
data['WAGE_Q'] = data['   '].astype(float)
data['M0_Q'] = data['  0 (  )'].str.replace(',', '.').astype(float)

#   
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['M0_Q'], y=data['WAGE_Q'])
plt.xlabel('  0 (  )')
plt.ylabel('    (WAGE_Q)')
plt.title('       ')
plt.show()

#    
correlation_coefficient, p_value = pearsonr(data['M0_Q'], data['WAGE_Q'])
print(" :", correlation_coefficient)
print("p-:", p_value)
```

```python
#Task_2
import statsmodels.api as sm

#    
y = data['WAGE_Q']
x = data['M0_Q']

#        
X = sm.add_constant(x)
model = sm.OLS(y, X).fit()

#   
print(model.summary())

#    
mean_relative_error = (abs((model.predict(X) - y) / y)).mean()
print("   :", mean_relative_error)
```

```python
#Task_3
from scipy.stats import shapiro

#  
residuals = model.resid

#  -
stat, p_value_shapiro = shapiro(residuals)
print("  -:", stat)
print("p-:", p_value_shapiro)
if p_value_shapiro > 0.05:
    print("  .")
else:
    print("    .")
```

```python
#Task_4
from statsmodels.stats.diagnostic import het_breuschpagan

#  -
bp_test = het_breuschpagan(residuals, X)
print(" -:", bp_test[0])
print("p-:", bp_test[1])
if bp_test[1] > 0.05:
    print(" .")
else:
    print(" .")

#      
plt.figure(figsize=(10, 6))
plt.scatter(model.fittedvalues, residuals, alpha=0.6)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel(" ")
plt.ylabel("")
plt.title("    ")
plt.show()
```

```python
#Task_5
import pandas as pd

#    
mean_x = x.mean()
new_x_value = 1.05 * mean_x

#  DataFrame  
new_X = pd.DataFrame({'const': [1], 'M0_Q': [new_x_value]})

#     
prediction = model.get_prediction(new_X)
prediction_summary = prediction.summary_frame(alpha=0.05)

#  
point_estimate = prediction_summary['mean'][0]
lower_bound = prediction_summary['mean_ci_lower'][0]
upper_bound = prediction_summary['mean_ci_upper'][0]
print("  :", point_estimate)
print("  (95%):", (lower_bound, upper_bound))

#      
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', label=' ')
plt.axvline(new_x_value, color='green', linestyle='--', label=f'105%   M0 ({new_x_value:.2f})')
plt.errorbar(new_x_value, point_estimate, yerr=[[point_estimate - lower_bound], [upper_bound - point_estimate]], fmt='o', color='red', label='   ')
plt.xlabel("  0 (  )")
plt.ylabel("    (WAGE_Q)")
plt.title("     105%    M0")
plt.legend()
plt.show()
```

```python
#Task_6
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

#     (   )
file_path = '_16_.xlsx'
data = pd.read_excel(file_path)

#       
data['M0_Q'] = data['  0 (  )'].str.replace(',', '.').astype(float)
data['WAGE_C_Q'] = data['   '].astype(float)

#    
y = data['M0_Q']  #   (  0)
x = data['WAGE_C_Q']  #   ( )

#   
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.7)
plt.xlabel("    (WAGE_C_Q)")
plt.ylabel("   0 (M0_Q)")
plt.title("       0")
plt.show()

#    
correlation_coefficient, p_value = pearsonr(x, y)
print(" :", correlation_coefficient)
print("p-:", p_value)

#    
X = sm.add_constant(x)  #  
model = sm.OLS(y, X).fit()

#   
print(model.summary())

#    
mean_relative_error = (abs((model.predict(X) - y) / y)).mean()
print("   :", mean_relative_error)
```
